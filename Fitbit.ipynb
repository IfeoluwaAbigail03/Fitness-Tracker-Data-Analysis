Task6:  group  the column totalsteps with Activitydates to find out the relationship between the two columns

```python
daily_activities_df['ActivityDate'] = pd.to_datetime(daily_activities_df['ActivityDate'])
Date_totalsteps=daily_activities_df.groupby('ActivityDate')['TotalSteps'].mean()
```

```python
Date_totalsteps_reset= Date_totalsteps.reset_index()
```

```python
daily_activities_df['ActivityDate'] = pd.to_datetime(daily_activities_df['ActivityDate'])
plt.figure(figsize=(12, 6))
sns.lineplot(x='ActivityDate', y='TotalSteps', data=daily_activities_df)
plt.scatter(daily_activities_df['ActivityDate'], daily_activities_df['TotalSteps'], marker='o', color='red', label='Total Steps')

#plt.title('Time Series: Total Steps')
plt.grid(True)
plt.show()
```

```python


Date_totalsteps_reset['DayOfWeek'] = pd.to_datetime(Date_totalsteps_reset['ActivityDate']).dt.dayofweek
def get_day_name(day):
    days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    return days[day]


Date_totalsteps_reset['DayName'] = Date_totalsteps_reset['DayOfWeek'].apply(get_day_name)
total_steps_by_day = Date_totalsteps_reset.groupby('DayName')['TotalSteps'].sum().reset_index()
sns.lineplot(x='DayName', y='TotalSteps', data=total_steps_by_day, marker='o', color='blue')
plt.xlabel('Day of the Week')
plt.ylabel('Total Steps')
#plt.title('Total Steps Taken by Day of the Week')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Specify the order of the x-axis labels
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

plt.grid(True)
plt.tight_layout()
plt.show()
```

Task 7: List numeric features to visualize, Loop through each numeric feature and create a histogram using Seaborn
 

```python
numeric_features = daily_activities_df[['TotalSteps', 'TotalDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryMinutes', 'Calories']]
numeric_features
```

```python
numeric_features = ['TotalSteps', 'TotalDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryMinutes', 'Calories']

for feature in numeric_features:
    plt.figure(figsize=(8, 5))
    sns.histplot(daily_activities_df[feature], bins=30, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

The plots above shows there is a decrease in the counts of of all the numeric values loops through ('TotalSteps', 'TotalDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryMinutes') except calories which shows a sharp rise and fall.

Task 8: Calculate the correlation matrix for the numeric features, Create a heatmap to visualize the correlation matrix with annotations

```python
correlation_matrix = daily_activities_df[numeric_features].corr()
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

The correlation maps above reveals there is high positive correlation between Totalsteps and Total distance, a positive correlation is shown between TotalSteps and Light active distance, TotalDistance and very Active distance, and a slightly positive correlation between calories and Total distance

Task 9: creat a list of of ActivityLevel, Loop through each activity level, create a bar plot for the median 'TotalDistance'  then create a bar plot to compare 'TotalDistance' by the current activity level

```python
ActivityLevel=['VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryMinutes']
for value in ActivityLevel:
  aggregated_data = daily_activities_df.groupby(value)['TotalDistance'].median().reset_index()
  plt.figure(figsize=(10, 6))
  sns.barplot(x=value, y='TotalDistance', data=aggregated_data)
  plt.title(f'Comparison of TotalDistance by {value}')
  plt.show()
```

The graph above shows increase in total distance with increase in light active distance and very active distance

Task 10: create a bar plot to compare 'Calories' by the activity level

```python
ActivityLevel=['TotalSteps', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryMinutes']
for value in ActivityLevel:
    Values=daily_activities_df.groupby(value)['Calories'].median().reset_index()
    plt.figure(figsize=(12,6))
    sns.barplot(x=value, y='Calories', data=Values)
    plt.title=(f'comparison of calories to {value}')
    plt.show()
```

Thereis no significant differences in activity level with calories
